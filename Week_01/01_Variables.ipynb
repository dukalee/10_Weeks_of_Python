### Variables

변수란 컴퓨터에 어떠한 값을 메모리에 저장하기 위해 활용합니다.

기본적인 숫자(정수, 실수), 문자열부터 리스트, 딕셔너리등 여러 데이터 타입을 저장할 수 있습니다.

변수를 만들 때에는 다음 규칙을 따라주어야 합니다.
* 왼쪽에 변수
* 오른쪽에 값
* 사이에 할당 연산자(=)
  * 프로그래밍에서 "="는 같다라는 의미가 아닙니다. 꼭 기억해 주세요.

```python
a = 25 # a라는 변수에 25라는 정수값을 저장
```

```python
print(a) # a를 출력
```

#### Variable Naming Rules

변수를 생성할때에는 다음과 같은 규칙을 따라주어야 합니다. 
* 변수명은 항상 알파벳(a ~ z, A ~ Z), 또는 언더바(_)로 시작해야 합니다.
* 변수명에는 언더바(_)를 제외한 특수문자가 허용되지 않습니다. 
* 변수명은 대소문자를 차별합니다. 
* 변수명에는 파이썬 키워드(keyword)를 활용할수 없습니다.

```python
age = 3 
_age = 4
age123 = 5
age_num = 6
```

#### Variable Naming Conventions

변수명은 아래의 관례를 따라주는 것을 권장합니다.
* snake_case형으로 변수명을 작성
* 변수명은 항상 담고자, 표현하고자 하는 값에 의미 있도록

image.png

```python
ExampleOfCamelCase = 3    # CamelCase
example_of_snake_case = 5 # snake_case
```

어떤 형태를 활용해도 정상적으로 작동합니다. 단, 파이썬에서는 snake_case의 활용을 권장합니다. (Java, C등은 CamelCase를 권장합니다.)

```python
a = 2021
b = "November"
c = 14
print(f"Today is {b} {c}, {a}")
```

```python
year = 2021
month = "November"
day = 14
print(f"Today is {month} {day}, {year}")
```

프로그램을 작성하다 보면 변수는 수없이 쌓이기 마련입니다. 

내가 필요한 변수를 손쉽게 찾고, 이해하기 위해 변수명은 항상 표현하고자 하는 값에 맞는 이름을 붙여주는 것을 권장합니다.
