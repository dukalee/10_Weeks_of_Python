### Operators

아래는 파이썬에서의 모든 연산자의 실행 우선순위입니다. 우리는 중요하게 다음의 순서만 알고 있으면 됩니다.

#### Assignment Operator

할당연산자 <code>=</code>는 프로그래밍 언어에서 변수에 값을 저장하기 위해 활용하는 연산자입니다.
* 우측에 값, 좌측에 변수가 대입됩니다.
* 변수의 이름은 : 
  * 알파벳으로 시작해야 하며,
  * 띄어쓰기나 특수문자가 존재할 수 없으며 (_ 제외), 
  * 이미 존재하는 명령어 또는 변수를 덮어씌우지 않도록 주의해야 하며, 
  * 담으려는 값을 최대한 표현할 수 있는 이름으로 지어주어야 합니다.


```python
a = 13 # a에 13을 저장
```

```python
print(a)
```

```python
print(a + 2)
```

##### Numerical Data

산술연산자는 일반적으로 수학에서 활용하는 연산자와 비슷합니다.

```python
print(3 + 3) # 더하기
```

```python
print(4 - 3) # 빼기
```

```python
print(4 * 4) # 곱하기
```

```python
print(10 / 3) # 나누기
```

##### Operator Precedence

산술 연산자의 우선순위는 다음과 같습니다.

* 괄호 ()
* 거듭제곱 **
* 단항(unary) +,-
* 곱셈, 나눗셈 *,/,//,%
* 이항(binary) +,- 

여기서 단항 +,- 는 부호 (-1, -3등)을 의미하고, 이항 +,-는 두 숫자의 덧셈, 뺄셈을 의미합니다.

```python
print(-(-10**2))
```

위 연산에 괄호를 쳐보면, (-(-(10)**2)) 가 됩니다.

#### Comparison Operators
---

비교 연산자는 두 값을 비교하여 참(True) 또는 거짓(False)로 결과를 나타냅니다.
* <code>></code> : 보다 크다
* <code><</code> : 보다 작다
* <code>>=</code> : 크거나 같다
* <code><=</code> : 작거나 같다
* <code>==</code> : 같다
* <code>!=</code> : 다르다

```python
print(3 > 5)
```

```python
print(3 + 5 <= 7 + 1)
```

```python
print(1 * 1 == 4 / 4) # int와 float의 비교라도 값만 동일하다면 ok
```

```python
print("Coding" == "coding") # 문자열도 같은지, 다른지는 비교 가능
```

##### Operator Precedence

* 모든 비교 연산자는 동일한 우선 순위를 가집니다.
* 위의 예에서 볼 수 있듯이, 비교 연산자의 우선순위는 산술연산자보다 낮습니다. (산술 연산을 먼저 실행하고, 비교 연산을 합니다.)

#### Logical Operators
---

논리 연산은 여러 부울값 (비교 연산의 결과)를 논리적으로 결합한 결과를 내기 위해 활용합니다. 
* <code>and</code> : 두 결과가 모두 "참"일때 참 
* <code>or</code> : 두 결과 중 하나만 "참"이여도 참
* <code>not</code> : 참이면 거짓, 거짓이면 참

```python
print(True and True)
print(True and False)
print(False and False)
```

```python
print(True or True)
print(True or False)
print(False or False)
```

```python
print(not True)
print(not False)
```

##### Chained Comparison

파이썬에서는 여러 비교를 한번에 할 수 있는 기능이 있습니다.

```python
a = 5 
print(1 <= a <= 10) # a는 1이상이고 10이하인가
```

위와 같이 여러 비교연산을 한번에 쓸 때에는, 하나의 비교 연산 사이에 "and"가 포함된 형태로 풀어쓸 수 있습니다.

```python
print(1 <= a and a <= 10)
```

그렇기 때문에 다음과 같은 오해가 없이 활용할수 있도록 주의해야 합니다.

```python
print(-1 < 0 == False)
print(-1 < 0 and 0 == False) # 0은 False로 판단됩니다.
```

```python
print((-1 < 0) == False) # True == False는 거짓입니다.
```

##### Operator Precedence

논리 연산자는 다음과 같은 우선순위를 가집니다.
* not
* and 
* or

```python
print(not 5 == 5 or 1 != 2)
```

```python
print(not(5 == 5 or 1 != 2))
```

#### Membership Operator

파이썬에는 집합형 데이터에서 "요소"를 포함하는지를 확인하는 특수한 멤버쉽 연산자가 있습니다.
* <code>in</code> : 특정 요소/값이 집합형 데이터(iterable)에 존재하는지를 확인합니다.

```python
alist = [1, 2, 3, 4, 5]
print(1 in alist)
```

```python
print(2 not in alist)
```

##### Operator Precedence

멤버쉽 연산은 비교 연산과 동일한 우선 순위를 가집니다.
* 즉, 만나는 순서대로 실행되기 때문에, 실수가 없도록 주의해야 합니다.

```python
y = "33"
print("3" in y == True) # 3이 y안에 포함된것이 참(진짜)인가?
```

위에서 쓴 코드를 풀어쓰면 다음과 같습니다.

```python
print("3" in y and y == True) # 3이 y안에 포함되어있고, y는 True와 같은가?
```

```python
print(y == True) # y는 부울타입 True가 아닙니다. (bool(y) == True 는 맞습니다.)
```

```python
print("3" in y) # "3"이 y안에 포함되어있습니다.
```

위에서 표현하려던 연산을 하려면 다음과 같이 코드를 작성해야 합니다.

```python
print(("3" in y) == True)
```
